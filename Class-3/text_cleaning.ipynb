## Lowercasing

Convert all text into lowercase for consistency.

```python
text = "AI is AMAZING"
clean_text = text.lower()
print(clean_text)
```

## Punctuation Removal

Remove unnecessary punctuation marks.

```python
import string 
text = "Hello !!! How are you ???"
for p in string.punctuation:
    text = text.replace(p, "")
print(text)
```

## Regular Expressions (Regex)

Regex helps clean text using patterns.

```python
import re
text = "Revenue 2024 = $500000"
clean = re.sub(r"\d+", "", text)
print(clean)
```

```python
import re
import string
text = "AI is AMAZING!!! Revenue in 2025 = $6789000 "
clean_text = text.lower()

for p in string.punctuation:
    text = text.replace(p, "")
text = re.sub(r"\d+", "", text)
print(text)
```
