# Installation of Required Library

```python
import nltk
```

# Download Pre-Trained Tokenizer Model

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

# Importing Sentence and Word Tokenization tools 

```python
from nltk.tokenize import word_tokenize
```

```python
from nltk.tokenize import sent_tokenize
```

# Given Text

```python
text = "I Love Artificial Intelligence. Gemini is Amazing"
```

# Displaying the Output

```python
print(sent_tokenize(text))
print(word_tokenize(text))
```
