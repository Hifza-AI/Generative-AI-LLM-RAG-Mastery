# TF-IDF

Convert text into weighted numerical vectors.

## Import TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

## Sample Documents

```python
documents = ['AI is boomimg everywhere',
             'AI is used in Finanance.',
             'I will definately become a Remotely AI Engineer']
```

## Create TF-IDF Object

```python
vectorizer = TfidfVectorizer()
```

## Learn Vocabulary and Calculate TF-IDF Scores

```python
x = vectorizer.fit_transform(documents)
```

## Displaying the Vocabulary

```python
print(vectorizer.get_feature_names_out())
```

## Displaying the Numerical Matrix

```python
print(x.toarray())
```

## Observation

TF-IDF improves Bag of Words by giving less importance to common words and more importance to unique words.
