## Stop Words

### Objective
Learn how to remove unnecessary words from text before further NLP processing.

```python
import nltk
from nltk.corpus import stopwords
```

## Step 2: Download Stop Words Dataset

Here we download the Stop Words dataset from NLTK.
This download is required only once.

```python
nltk.download('stopwords')
```

## Step 3: Load English Stop Words

Load all English stop words and convert them into a Python Set for faster searching.

```python
stop_words = set(stopwords.words('english'))
```

## Step 4: Create a Sample Sentence

Create a sample sentence that will be used to demonstrate Stop Words Removal.

```python
sentence = "I love Generative AI and I am Learning NLP"
```

## Step 5: Split Sentence into Words

Split the sentence into individual words so that each word can be processed separately.

```python
words = sentence.split()
```

## Step 6: Remove Stop Words

Iterate through each word, compare it with the Stop Words list, and keep only the meaningful words.

```python
filtered_words = []
for word in words:
    if word.lower() not in stop_words:
        filtered_words.append(word)
```

## Step 7: Display the Results

Print both the original words and the filtered words after removing Stop Words.

```python
print("Original Words:")
print(words)

print("/nAfter Removing Stop Words:")
print(filtered_words)
```

## Observation

Original Sentence:
I love Artificial Intelligence and I am learning NLP

After Removing Stop Words:
love Artificial Intelligence learning NLP

The words "I", "and", and "am" were removed because they are Stop Words.

Removing Stop Words helps NLP models focus on meaningful words while reducing unnecessary information.

## Real World Connection

Stop Words Removal is commonly used in traditional NLP tasks such as:

- Text Classification
- Search Engines
- Information Retrieval
- TF-IDF
- Bag of Words

In modern LLMs like GPT and BERT, Stop Words are not always removed because they help preserve sentence meaning and context.

In RAG systems, Stop Words Removal depends on the use case and retrieval strategy.
