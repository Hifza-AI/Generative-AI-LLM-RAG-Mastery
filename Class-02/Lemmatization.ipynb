## Lemmatization

## Objective
Learn how Lemmatization converts words into their dictionary (real) form.

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer
```

## Step 1: Create a Lemmatizer

Here we create an object of the WordNetLemmatizer class.
This object will perform lemmatization on words.

```python
lemmatizer = WordNetLemmatizer()
```

## Step 2: Create Sample Words

These words will be used to understand how lemmatization works.

```python
words = ["playing", "played", "running", "studies"]
```

## Step 3: Apply Lemmatization

Loop through each word and convert it into its base form.

```python
for word in words:
  print(word, "->", lemmatizer.lemmatize(word))
```

## Step 4: Verb Lemmatization

By default, lemmatizer assumes words are nouns.

Here we tell it that the words are verbs using:

pos='v'

```python
for word in words:
  print(word, "->", lemmatizer.lemmatize(word, pos='v'))
```

# Observation

Without POS:

Playing -> Playing
Played -> Played
running -> run
studies -> study

With POS='v':

Playing -> Play
Played -> Play
running -> run
studies -> study

Conclusion:

Lemmatization returns meaningful dictionary words.

It is more accurate than stemming.

Most modern NLP systems (BERT, GPT, RAG, LLMs) prefer lemmatization over stemming whenever linguistic accuracy is important.
