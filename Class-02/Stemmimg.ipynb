## Stemming

Objective:
Learn how Stemming converts words into their root form.

```python
import nltk
from nltk.stem import PorterStemmer
```

## Create Stemmer

Now we create a PorterStemmer object which will perform stemming.

```python
stemmer = PorterStemmer()
```

## Input Words

These words will be used as input for stemming.

```python
words = ["playing", "play", "running", "studies"]
```

## Apply Stemming

Apply the stem() function to each word.

```python
for word in words:
    print(word, "->" , stemmer.stem(word))
```

## Observation

playing → play

studies → studi

running → run

Stemming removes suffixes without checking dictionary meaning.
