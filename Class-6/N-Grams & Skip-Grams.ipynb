N-Grams and Skip-Grams

## Import Required Library

Import CountVectorizer from scikit-learn Library.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

## Create Sample Sentences

We will use two example sentences.

```python
sentences = [
    "I Love AI",
    "AI is amazing"
]
```

## Create Bigram Vectorizer

Generate only bigrams using CountVectorizer.

```python
vectorizer = CountVectorizer(ngram_range=(2,2))
x = vectorizer.fit_transform(sentences)
```

## Display Vocabulary

Show generated bigram features.

```python
print(vectorizer.get_feature_names_out())
```

## Convert into Matrix

Display numerical representation.

```python
print(x.toarray())
```

## Observation

N-Grams improve local context compared to Bag of Words.

Modern RAG systems usually use embeddings instead of N-Grams.
