## Import CountVectorizer

CountVectorizer converts text into numerical vectors using the Bag of Words technique.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

## Creating Sample Dataset

We define two simple sentences for understanding Bag of Words.

```python
sentences = [
    "I Love AI",
    "I Love Python"
]
```

## Creating CountVectorizer Object

The object learns the vocabulary from the given text.

```python
vectorizer = CountVectorizer()
```

## Learning Vocabulary and Creating Vectors

The fit_transform() method first learns the vocabulary and then converts the text into numerical vectors.

```python
x = vectorizer.fit_transform(sentences)
```

## Displaying the Vocabulary

This shows all unique words learned by the CountVectorizer.

```python
print(vectorizer.get_feature_names_out())
```

## Displaying the Numerical Matrix

The matrix shows how many times each vocabulary word appears in each sentence.

```python
print(x.toarray())
```
